# Latihan Sklearn Logistic Regression

## Tujuan

Pada latihan ini kita akan menggunakan logistic regression untuk memprediksi apakah seseorang akan membeli setelah melihat iklan sebuah produk.

## Tahapan Latihan

Tahapan yang dilalui dalam latihan kali ini adalah sebagai berikut:

1. Ubah dataset menjadi Dataframe.
2. Hapus kolom 'User ID'.
3. Pisahkan atribut dan label.
4. Latih model Logistic Regression.
5. Evaluasi akurasi model.

Dataset: https://www.kaggle.com/datasets/dragonheir/logistic-regression/


```
import pandas as pd

df = pd.read_csv('https://github.com/sabirinID/IDCamp/raw/main/Dicoding/Machine%20Learning%20untuk%20Pemula/Dataset/Social_Network_Ads.csv')
```

```
df.head()
```

```
df.info()
```

```
# Menghapus kolom yang tidak diperlukan
data = df.drop(columns=['User ID'])

# Menjalankan proses one-hot encoding dengan pd.get_dummies()
data = pd.get_dummies(data)
data
```

```
# Memisahkan atribut dan label
predictions = ['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']

X = data[predictions]
y = data['Purchased']
```

```
# Melakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns= X.columns)
scaled_data.head()
```

```
from sklearn.model_selection import train_test_split

# Membagi data menjadi train dan test untuk setiap atribut dan label
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
print(len(y_test))
```

```
from sklearn import linear_model

# Melatih model dengan fungsi fit
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```
# Menguji akurasi model
model.score(X_test, y_test)
```
